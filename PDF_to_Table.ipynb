```python
!pip install easyocr -q
import easyocr
reader=easyocr.Reader(['en'])
import PIL
from PIL import ImageDraw
```

```python
jpgs =  os.listdir('/content/drive/MyDrive/700t')
jpgs =  ['/content/drive/MyDrive/700t/'+i  for i in jpgs]
images=[]
for i in jpgs:
  if i.find('.jpg')!=-1:
    images.append(i)
```

```python
images
```

```python
def extract_table(img_path):
  im=PIL.Image.open(img_path)
  bounds=reader.readtext(img_path)

  def draw_boxes(image,bounds,color = 'yellow',width = 3):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
      p0,p1,p2,p3 = bound[0]
      draw.line([*p0,*p1,*p2,*p3], fill = color,width = 2)
    return image

  draw_boxes(im,bounds)

  text=''
  for i in range(len(bounds)):
    text=text+bounds[i][1]+ '\n'


  name=[]
  date=[]
  for i in range(len(bounds)):
   if re.match('SEC-F', bounds[i][1]):
     break
   if re.match('[A-z][A-z]', bounds[i][1].replace(" ", "")):
     if bounds[i][1] not in ['PX','pX','px','Px']:
      name.append(bounds[i][1])
      date.append(bounds[i+2][1].replace(" ", ""))  

  Data=pd.DataFrame(zip(name,date), columns=['Name', 'Date'] )
  Data['Letter'] = Data.index.to_series().apply(lambda x: chr(ord('a') + x)).str.upper()
  Data['Fisch No']=bounds[0][1].replace(" ",'')
  return(Data[['Letter', 'Name', 'Date', 'Fisch No']])
```

```python
Table=[]
for i in images:
  Table.append(extract_table(i))
```
